```python
import requests 
import bs4
import pandas as pd
import numpy as np
import json
import matplotlib 
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
URL = 'https://en.wikipedia.org/wiki/Apple'
```

```python
response = requests.get(URL)
if response.status_code == 200:
    print ("Get Successfull")
else:
    print ("Get Failed")
```

```python
resp_soap = bs4.BeautifulSoup(response.text)
```

```python
resp_soap.title
```

```python
# instantiate a word cloud object
response_wc = WordCloud(background_color='white',max_words=1000)

# generate the word cloud
wc_apple = response_wc.generate(resp_soap.get_text())
```

```python
fig = plt.figure()
fig.set_figwidth(16)
fig.set_figheight(6)
plt.imshow(response_wc, interpolation='bilinear')

plt.axis('off')
plt.show()
```
